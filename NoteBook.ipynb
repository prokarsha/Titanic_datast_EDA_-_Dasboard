## Dataset Visualization

```python
# Visualization of survival distribution
sns.countplot(x='Survived', data=df)
plt.title('Survival Distribution')
plt.show()
```

### The countplot provides a clear visualization of the distribution of survival outcomes, showing the number of passengers who survived (1) and those who did not (0). It visualise a quick overview of the overall survival rate in the Titanic dataset.

```python
# Visualization of survival based on Pclass
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival based on Pclass')
plt.show()
```

### This countplot depicts the distribution of survival outcomes categorized by passenger class (Pclass). It helps to understand whether there are variations in survival rates among different ticket classes, offering insights into potential socio-economic factors affecting survival.

```python
# Age distribution
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

### The histogram displays the distribution of passenger ages, revealing the concentration of individuals within specific age ranges. It aids in understanding the age demographics of Titanic passengers and identifying any noticeable patterns in age distribution.

```python
# Gender distribution
sns.countplot(x='Sex', data=df)
plt.title('Gender Distribution')
plt.show()
```

### The countplot represents the distribution of passengers based on gender, providing a visual breakdown of the number of males and females aboard the Titanic. It visualise insights into the gender composition of the dataset.

```python
# Embarked distribution
sns.countplot(x='Embarked', data=df)
plt.title('Embarked Distribution')
plt.show()
```

### This countplot illustrates the distribution of passengers based on their port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton). It helps to understand the embarkation patterns and the relative frequency of passengers from each port.

```python
# Correlation matrix
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

### The correlation heatmap visualizes the correlation matrix of numerical variables in the dataset. Brighter colors indicate stronger correlations, and it helps identify relationships between features. In this dataset, the heatmap could reveal potential connections, such as correlations between ticket class (Pclass) and fare, or age and the number of siblings/spouses (SibSp). This assists in understanding how different variables are interrelated and might impact survival.

```python
# Pairplot
sns.pairplot(df, hue='Survived', diag_kind='kde')
plt.suptitle('Pairplot of Titanic Dataset', y=1.02)
plt.show()
```

### The pairplot illustrates relationships between numerical variables such as Age, Fare, SibSp, Parch, and their distribution with respect to survival status. It helps to identify potential patterns and trends in the Titanic dataset.

```python
# Jointplot for Age and Fare
sns.jointplot(x='Age', y='Fare', data=df, kind='scatter')
plt.suptitle('Jointplot of Age and Fare', y=1.02)
plt.show()
```

### The jointplot displays a scatter plot of Age and Fare, offering insights into the distribution of passengers' ages and ticket fares. It helps to visualise the concentration of points and potential clusters within the data.

```python
# Bivariate plot for Age and Fare
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Bivariate Plot of Age and Fare with Survival')
plt.show()
```

### The bivariate scatterplot highlights the relationship between Age and Fare, with different colors indicating whether passengers survived or not. It allows for a quick assessment of any discernible patterns or differences in the distribution of Age and Fare between survivors and non-survivors.

```python
# Jointplot for Age and Fare with hue indicating survival
sns.jointplot(x='Age', y='Fare', data=df, hue='Survived', kind='scatter')
plt.suptitle('Jointplot of Age and Fare with Survival', y=1.02)
plt.show()
```

### This jointplot illustrates the relationship between Age and Fare, with different colors indicating whether passengers survived or not. It helps to visualise the joint distribution of these two variables and provides insights into potential patterns or concentrations related to survival

```python
# Bivariate plot for Pclass and Fare with hue indicating survival
sns.scatterplot(x='Pclass', y='Fare', hue='Survived', data=df)
plt.title('Bivariate Plot of Pclass and Fare with Survival')
plt.show()
```

### The bivariate scatterplot explores the relationship between Passenger Class (Pclass) and Fare, with survival status distinguished by different colors. It allows for a quick assessment of any correlations between ticket class, fare paid, and the likelihood of survival among passengers

```python
# Bivariate plot for SibSp and Parch with hue indicating survival
sns.scatterplot(x='SibSp', y='Parch', hue='Survived', data=df)
plt.title('Bivariate Plot of SibSp and Parch with Survival')
plt.show()
```

### This scatterplot depicts the relationship between the number of siblings/spouses (SibSp) and the number of parents/children (Parch), with survival status indicated by different colors. It helps to visualise the potential influence of family size on survival and identifies any notable patterns or distinctions between survivors and non-survivors.

## Subplot

```python
# Create a 2x2 subplot grid
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Subplot 1: Pairplot
sns.set(style="whitegrid")
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Pairplot of Titanic Dataset')

# Subplot 2: Jointplot for Age and Fare
sns.scatterplot(x='Age', y='Fare', data=df, hue='Survived', ax=axes[0, 1])
axes[0, 1].set_title('Jointplot of Age and Fare with Survival')

# Subplot 3: Countplot for Survival based on Pclass
sns.countplot(x='Pclass', hue='Survived', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Survival based on Pclass')

# Subplot 4: Bivariate Plot for SibSp and Parch with Survival
sns.scatterplot(x='SibSp', y='Parch', hue='Survived', data=df, ax=axes[1, 1])
axes[1, 1].set_title('Bivariate Plot of SibSp and Parch with Survival')

# Adjust layout
plt.tight_layout()
plt.show()
```

### This subplot depicts the different visualisation technique with survival status indicated by different colors. It helps to visualise the potential influence of the datasets to identify the pattern.

## Dashboard

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
```

```python
# Initialize the Dash app
app = dash.Dash(__name__)

# Define layout
app.layout = html.Div([
    # Dropdown menu for variable selection
    dcc.Dropdown(
        id='variable-dropdown',
        options=[
            {'label': 'Age', 'value': 'Age'},
            {'label': 'Fare', 'value': 'Fare'},
            # Add more variables as needed
        ],
        value='Age',  # Default selected variable
        style={'width': '50%'}
    ),
    
    # Scatter plot based on the selected variable
    dcc.Graph(id='scatter-plot')
])

# Define callback to update scatter plot based on dropdown selection
@app.callback(
    Output('scatter-plot', 'figure'),
    [Input('variable-dropdown', 'value')]
)
def update_scatter_plot(selected_variable):
    fig = px.scatter(df, x=selected_variable, y='Fare', color='Survived', 
                     title=f'Scatter Plot of {selected_variable} vs Fare with Survival')
    return fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
